```python
import seaborn as sns
import pandas as pd
```

```python
df_ad = pd.read_csv('https://raw.githubusercontent.com/chiharuishida/Tutorialdata/main/ad_dataset_withheaders.csv')
```

```python
# First 10 rows of dataset
df_ad.head(10)
```

```python
# Descriptive statistics for all the continous variables in the dataset
df_ad.describe()
```

```python
# Provides information about each variable in the dataset
df_ad.info()
```

```python
df_ad['Gender'].value_counts()
```

```python
df_ad['AdClicking'].value_counts()
```

```python
df_ad[['Gender','AdClicking']].value_counts()
```

```python
# Correlation matrix
df_ad.corr()
```

```python
# Heat map of correlation matrix
sns.heatmap(df_ad.corr(), annot=True)
```

```python
# Changing the color of the heatmap
sns.heatmap(df_ad.corr(), cmap="YlGnBu")
```

```python
# Provides pairplot between the continous varialbles
sns.pairplot(df_ad)
```

```python
# Scatterplot between "Age" and "DailyTime"
sns.scatterplot(x="Age", y="DailyTime", data=df_ad)
```

```python
# Fitting trendline for Scatterplot between "Age" and "DailyTime"
sns.lmplot(x="Age", y="DailyTime", data=df_ad)
```

```python
#adding color-coding by Gender by adding the hue statement
sns.scatterplot(x="Age", y="DailyTime", hue="Gender", data=df_ad)
```

```python
#Fitting lines for the two gender categories
sns.lmplot(x="Age", y="DailyTime", hue='Gender', data=df_ad)
```

```python
# Bar chart sorted by "Gender" which is a categorical variable
sns.countplot(x='Gender', data=df_ad)
```

```python
#changing color to all red by adding "color='b'"
sns.countplot(x='Gender', data=df_ad, color='b')
```

```python
# groupby() gender is used to split the data in dataframe into two groups
df_ad_gender=df_ad.groupby("Gender")
df_ad_gender.mean()


```

```python
# Boxplot of AdClicking vs Age
sns.boxplot(x="AdClicking", y="Age", data=df_ad)
```

```python
# Boxplot of AdClicking vs Age and color code for "Gender"
sns.boxplot(x="AdClicking", y="Age", hue="Gender", data=df_ad)
```

```python
# Top 3 DailyTime spent
df_ad_top3 = df_ad.nlargest(3, "DailyTime")
df_ad_top3
```
